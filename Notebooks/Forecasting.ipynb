```python
# 1. Install Prophet (Colab needs this)
!pip install prophet

import pandas as pd
from prophet import Prophet

# 2. Load the uploaded file
df = pd.read_csv('Superstore.csv')

# 3. Prepare Data
df['ds'] = pd.to_datetime(df['Order Date'], dayfirst=True, errors='coerce')
df['y'] = df['Sales']
daily_sales = df.groupby('ds')['y'].sum().reset_index()

# 4. Run Model
m = Prophet(interval_width=0.95, daily_seasonality=True)
m.fit(daily_sales)

# 5. Predict
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)

# 6. Save Result
final_data = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
final_data.columns = ['Order Date', 'Forecast', 'Lower Bound', 'Upper Bound']
final_data.to_csv('Sales_Forecast_Output.csv', index=False)

print("SUCCESS! Download your file from the left sidebar.")
```
